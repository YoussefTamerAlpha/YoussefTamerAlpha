```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

# Support: 

Quantify how many times an item or an itemset appear in a set of transactions.

In other words, support quantifies the frequency of an itemset.

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

# Confidence: 
After buying item X, what’s the probability of buying item Y?

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

# **Lift**: 
What’s the probability of buying items X and Y together rather than just buying item X.

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

```python
import pandas as pd

data = pd.read_csv('/content/Market_Basket_Optimisation.csv')
data
```

```python
data.info()
```

```python
print(data.isnull().sum())
```

```python
data.fillna('',axis=1,inplace=True)
data.head()
```

```python
data_list = data.to_numpy().tolist()
data_list
dataset = list()
for i in range(len(data_list)) :
    item = list()
    for j in data_list[i] :
        if pd.notna(j):             # DataFrame.notna() >>> Detect existing (non-missing) values.
            item.append(j)
    dataset.append(item)
```

```python
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
```

```python
te_array = te.fit(dataset).transform(dataset)
te_array
```

```python
te.columns_     # it got all the Dataframe unique values
```

```python
final_df = pd.DataFrame(te_array,columns=te.columns_)
final_df
```

```python
# remove the first column as it does not contain any infomation
final_df.drop(columns=[''],axis=1,inplace=True)
final_df
```

## Support:

 Quantify how many times an item or an itemset appear in a set of transactions.

In other words, support quantifies the frequency of an itemset.

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets=apriori(final_df, min_support=0.01, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.2) # associate itemsets with confidence over 20%.
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```
