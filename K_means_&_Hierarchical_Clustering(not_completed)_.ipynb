```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit(data)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
```

```python
X = data.iloc[:,:].values
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(random_state=42)
```

```python
from yellowbrick.cluster import KElbowVisualizer

visualizer = KElbowVisualizer(kmeans, k=(2,20), metric = 'distortion', timings = False)
visualizer.fit(X)
visualizer.show()
```

```python
kmeans = KMeans(n_clusters = 8, init = 'k-means++', random_state = 42)
label = kmeans.fit_predict(data)
x = label.tolist()
```

‘k-means++’ : selects initial cluster centroids

#Heirarchical Clustering


```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("cardholders  Dendrogram")

# Selecting Annual Income and Spending Scores by index

clusters = shc.linkage(data, 
            method='ward', 
            metric="euclidean")
shc.dendrogram(clusters)
plt.show()
```

# The optimal no. of clusters from the dendrogram are 8 also

The best K value is ***8***
