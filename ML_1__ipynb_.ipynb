```python
def label (row):
  if row['yr_built'] < 1950 :
    return 'Very_Ancient'
  elif row['yr_built'] < 1999 :
    return 'Ancient'
  else :
    return 'New'
```

```python
data.apply (lambda row: (row), axis=1)
```

```python
def label_bass (row):
  if row['sqft_basement'] < 300 :
    return 'wide'
  elif row['sqft_basement'] > 300 :
    return 'very_wide'
  elif row['sqft_basement'] ==0:
    return 'NaN'
```

```python
def label_renovated (row):
  if row['yr_renovated'] > 0  :
    return 1
  elif row['yr_renovated'] == 0 :
    return 0
  else:
    return 'NaN'
```

```python
data['HouseCondition'] = data.apply (lambda row: label(row), axis=1)
```

```python
data['sqft_basement'] = data.apply (lambda row: label_bass(row), axis=1)
```

```python
data['yr_renovated'] = data.apply (lambda row: label_renovated(row), axis=1)
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data['date'] = data['date'].dt.strftime('%Y')
# data['date'] = data['date'].dt.strftime('%Y')
```

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
data['HouseCondition']=encoder.fit_transform(data['HouseCondition'])
data['sqft_basement']=encoder.fit_transform(data['sqft_basement'])
```

```python
import  sklearn
from sklearn.model_selection import train_test_split
```

```python
y = data[['price']]
x = data[['sqft_living']]
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=40)
# it will learn on about 80% of the data , and test the rest
```

```python
# 4. Apply linear regression to your training set. 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error 
```

```python
from sklearn import metrics
```

```python
model=LinearRegression()
model.fit(x_train, y_train)
predicted=model.predict(x_test)
```

```python
x.shape
```

```python
y.shape
```

```python
print("MSE",mean_squared_error(y_test,predicted))
print("R squared",metrics.r2_score(y_test,predicted))

# ValueError: Found input variables with inconsistent numbers of samples: [4323, 17290]   when i put y_test
```
