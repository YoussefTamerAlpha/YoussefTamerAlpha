```python
#When loading data from potentially unstructured data sets,
# it can be useful to remove spaces and 
#lowercase all column names using a lambda (anonymous) function:
import pandas as pd

data = pd.read_csv("/content/datafile (2).csv")
data.rename(columns=lambda x: x.lower().replace(' ', '_'))


data.head()
```

```python
# Output data to a CSV file
# Typically, I don't want row numbers in my output file, hence index=False.
# To avoid character issues, I typically use utf8 encoding for input/output.

data.to_csv("output_filename.csv", index=False, encoding='utf8')

# Output data to an Excel file.
# For the excel output to work, you may need to install the "xlsxwriter" package.

data.to_csv("output_excel_file.xlsx", sheet_name="Sheet 1", index=False)
```

```python
df = pd.read_csv("/content/datafile (1).csv")

df.head()
```

```python
a = df['State'].value_counts()
a.head()
```

```python
plt.figure(figsize=(20,6))
plt.axis([0,10 , 0,10])
plt.title("Bar Chart")
plt.xlabel("Crop")
plt.ylabel("Amount")
plt.bar(a.index,a.values)
```

```python
total = df.shape[0]

perc = [a.iloc[0]/total,
        a.iloc[1]/total,
        a.iloc[2]/total,
        a.iloc[3]/total]
perc
```

```python
plt.title('PIE chart')
plt.axis('equal')
explode = (0.1,0.2,0,0.1)
plt.pie(perc,  autopct='%1.2f%%' ,startangle=90)           #ناقصها حاجة
plt.show()
```
